```python
resumen=df.groupby(['clasificacion']).size()
resumen
```

```python
def estado(x):
    if x.find('Fallecido')>0: 
        return "Fallecido"
    elif x.find('confirmado'):
        return "Confirmado"
    else return "Otro"    

df['clasificacion']=df.apply(lambda x: estado(x['clasificacion']),axis=1)
```
