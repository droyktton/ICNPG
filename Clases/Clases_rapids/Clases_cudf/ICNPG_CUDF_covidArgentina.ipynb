```
import cudf

# read CSV from memory
cuda_df = cudf.read_csv('/content/Covid19Casos.csv')
```

```

cuda_df.head()
```

selecciono columnas

```
cuda_df[['sexo','edad','carga_provincia_nombre']]
```

un nuevo data frame filtrando con condiciones, porteños menores de 35

```
cuda_df['edad'] = cudf.to_numeric(cuda_df['edad'], errors='coerce')
caba = cuda_df[(cuda_df["residencia_provincia_nombre"] == 'CABA') & (cuda_df["edad"] < 35)]
```

```
caba.head()
```

```
male_20_30 = cuda_df[(cuda_df['sexo'] == 'M') & (cuda_df['edad'] >= 20) & (cuda_df['edad'] <= 30)]
print(f"Number of male cases between 20 and 30 years old: {len(male_20_30)}")
```
