```python
import numpy as np
import cupy as cp
import time
from timeit import default_timer as timer
import cupyx.scipy.sparse.linalg as spla
import cupyx  

m=10
n=10

rala=cp.sparse.random(m, n, density=0.4, format='csr')

print(rala)

b=cp.arange(m)
x=cupyx.scipy.sparse.linalg.spsolve(rala,b)

A=cupyx.scipy.sparse.csr_matrix.toarray(rala)
print(cp.matmul(A,x))
```
