# CUDSS Install

No es parte del toolkit aun.

```python
%%bash

wget https://developer.download.nvidia.com/compute/cudss/redist/libcudss/linux-x86_64/libcudss-linux-x86_64-0.7.1.4_cuda12-archive.tar.xz
```

```python
%%bash

tar xvf /content/libcudss-linux-x86_64-0.7.1.4_cuda12-archive.tar.xz
```

```python
%%bash

CUDSS_DIR=/content/libcudss-linux-x86_64-0.7.1.4_cuda12-archive
export CUDSS_DIR

CTK_DIR=/usr/local/cuda-12
export CTK_DIR

export LD_LIBRARY_PATH=${CUDSS_DIR}/lib:${CTK_DIR}/lib64:${LD_LIBRARY_PATH}

cd $CUDSS_DIR/examples
#ls

nvcc -arch=sm_75 cudss_simple.cpp -I${CUDSS_DIR}/include -L${CUDSS_DIR}/lib -lcudss -lcublas -o cudss_simple

./cudss_simple
#ls
#cp $CUDADSS_DIR/examples/cudss_simple.cpp cudss_simple.cu
#nvcc -arch=sm_75 cudss_simple.cu -I${CUDSS_DIR}/include -L${CUDSS_DIR}/lib -lcudss -lcublas -o cudss_simple

```
